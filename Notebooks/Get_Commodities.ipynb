```python
import yfinance as yf
from datetime import datetime

# Define o ticker
ticker_symbol = "MSFT"

# Cria o objeto ticker
ticker = yf.Ticker(ticker_symbol)

# Obtenha dados históricos de mercado dos últimos 30 dias
historical_data = ticker.history(period="1d", interval="1m") [['Close']]

historical_data = historical_data.rename(columns={'Close': 'preco'})
historical_data['ativo'] = ticker_symbol
historical_data['moeda'] = 'USD'
historical_data['hora_request'] = datetime.now()
historical_data_Final = historical_data[['preco','ativo','moeda','hora_request']]

historical_data_Final
# Exibir um resumo dos dados obtidos em um dataframe
#print(f"Resumo de dados históricos para {ticker_symbol}:")
#print(historical_data[['Open', 'High', 'Low', 'Close', 'Volume']])
```
